**Introduction**

In this notebook, we will explore the basic components behind the Perceptron Model. The perceptron serves as a fundamental unit of logic in artificial neural networks. It is primarily designed for binary classification tasks, distinguishing between two possible outcomes (e.g., 1 or 0, yes or no, etc.). While the perceptron has its limitations, its simplicity provides a strong foundation for understanding more complex models.

**Preparing Perceptron Inputs & Weights**

We have a set of inputs, *x0, x1, x2*, each with a specific value. Similarly, we have a set of weights, *w0, w1, w2*, each assigned a corresponding value. These weights dictate the influence each input value has on the overall outcome.

```python
data = {"Inputs" : [0.6, 0.4, 0.4],
        "Weights" : [0.4, 0.3, 0.3]}
```

**Multiplying Corresponding Inputs & Weights & Summing Corresponding Products**

The input *x0*​, with a weight of 0.4, is the most significant since its weight is greater than those of *x1*​ and *x2*​. To reach a logical conclusion, we multiply each input by its corresponding weight and then calculate the sum of these products.

```python
corresponding_products = [input_value * weight_value for input_value, weight_value in zip(data["Inputs"], data["Weights"])]
print("Corresponding Products:",corresponding_products)
```

Take the array of corresponding products and find the sum of it. In this case the sum of the corresponding products is 0.48

```python
sum_of_products = sum(corresponding_products)
print("Sum of Corresponding Products:", sum_of_products)
```

**Applied Activation Function**

Next, we check if the sum exceeds a certain threshold, also known as the bias. To evaluate whether the threshold has been met, we use an activation function called the step function. Our bias/threshold value will be 0.5.

$$
f(x) =
\begin{cases}
1 & \text{if } x > \text{0.5}, \\
0 & \text{otherwise}.
\end{cases}
$$


```python
def step_function(x, bias):
  return 1 if x > bias else 0

step_function_result = step_function(sum_of_products, 0.5)
print("Step Function Result:", step_function_result)
```
